```python
import numpy as np
import pandas as pd
import os
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(7,4))

def R2(x, y):
    fit = np.polyfit(x, np.log(y), 1)
    yf = fit[0]*x + fit[1]
    result = 1 - (sum((np.log(y)-yf)**2)) / (sum((np.log(y)-np.mean(np.log(y)))**2))
    return round(result, 4)
```

```python
# CU 60 (s)
b_cu = 23.8/60
Id_cu = (np.array([81, 69.2, 64.6, 63.8, 61.2]))/60 - b_cu
I0_cu = 81/60 - b_cu
d_cu = np.array([0, 1.01, 2.02, 3.02, 4.02])
dsm_cu = np.linspace(min(d_cu), max(d_cu), 1000)

fit_cu = np.polyfit(d_cu, np.log(Id_cu), 1)
yf_cu = fit_cu[0]*(dsm_cu) + fit_cu[1]

plt.scatter(d_cu, Id_cu, color='blue', label="Dữ liệu thực nghiệm")
plt.plot(dsm_cu, np.exp(yf_cu), label="Hàm hồi quy", c="dodgerblue")

delta_cu = np.sqrt( I0_cu*np.exp(2*fit_cu[0]*d_cu)/60 + b_cu/60 )
plt.errorbar(d_cu, Id_cu, yerr=delta_cu, linestyle='none', capsize=5, label="Sai số")

plt.title(r"Sự suy giảm của $Co^{60}$" + "\n" "khi qua vật liệu Cu", fontsize=13)
plt.xlabel(r"$d \ (mm)$", fontsize=13)
plt.ylabel("Số đếm/s", fontsize=13)
plt.legend(fontsize=11)
plt.text(0.5, 1.05, r"$y=0,89e^{-0,10x}$", fontsize=13, color="blue")
plt.text(0.5, 1.05-0.05, r"$R^2 = 0,8628$", fontsize=13, color="blue")
R2_cu = R2(d_cu, Id_cu)
print(R2_cu)

#plt.savefig('cu.png', dpi=300)
```

```python
print(sum(d_cu**2), sum(d_cu), sum(d_cu*np.log(Id_cu)))
print(sum(d_cu), sum(np.log(Id_cu)))
```

```python
# Al
d_al = np.array([0, 20.07, 45.75, 70.75])
I0_al = 12729
Id_al = np.array([12729,8594,4972,2966])
dsm_al = np.linspace(min(d_al), max(d_al), 1000)

fit_al = np.polyfit(d_al, np.log(Id_al), 1)
yf_al = fit_al[0]*(dsm_al) + fit_al[1]

plt.scatter(d_al, Id_al, color='red', label="Dữ liệu thực nghiệm")
plt.plot(dsm_al, np.exp(yf_al), label="Hàm hồi quy", c="crimson")

delta_al = np.sqrt(Id_al)
plt.errorbar(d_al, Id_al, yerr=delta_al, linestyle='none', capsize=5, label="Sai số")

R2_al = R2(d_al, Id_al)
print(R2_al)

plt.title(r"Sự suy giảm của $Cs^{137}$" + "\n" "khi qua vật liệu Al", fontsize=13)
plt.xlabel(r"$d \ (mm)$", fontsize=13)
plt.ylabel("Số đếm", fontsize=13)
plt.legend(fontsize=11)
plt.text(2, 6000, r"$y=13835,88e^{-0,02x}$", fontsize=13, color="crimson")
plt.text(2, 6000-1000, r"$R^2 = 0,9998$", fontsize=13, color="crimson")
#plt.savefig('al.png', dpi=300)
```
