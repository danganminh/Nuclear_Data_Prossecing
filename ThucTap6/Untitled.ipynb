```python
import os
import pandas as pd
import numpy as np
```

```python
data = pd.read_excel('table clear.xlsx')
display(data)
```

```python
mp = np.mean(data['mp'][0:10])
m1 = np.mean(data['m1'][0:10])
m12 = np.mean(data['m12'][0:10])
m2 = np.mean(data['m2'][0:10])
print(mp)
```

```python
# dead times
x = m1*m2 - mp*m12
y = m1*m2*(m12 + mp) - mp*m12*(m1 + m2)
z = ( y*(m1 + m2 - m12 - mp) )/(x**2)

tao = ( x*(1 - np.sqrt(1-z)) )/y
round(tao, 6)
```

```python
data_n = {
    'n1': data['m1'][0:10] / (1 - data['m1'][0:10]*tao),
    'n2': data['m2'][0:10] / (1 - data['m2'][0:10]*tao),
    'n12': data['m12'][0:10] / (1 - data['m12'][0:10]*tao),
}

df = pd.DataFrame(data_n)
display(df)

#df.to_excel('datan.xlsx')
```

```python
# const geometry G
a = 4.2
r = 3.31/2

G = ( 1 - a/(np.sqrt(a**2 + r**2)) )/2
round(G, 4)
```

```python
# efficiency Fe
n = np.mean(df['n1'][0:])
v = 2
A = 10**(-6) * 3.7 * 10**(10)

Fe = n/(A*G*v)

round(Fe*100, 4)
```
