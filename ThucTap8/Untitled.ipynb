```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import os
```

```python
data = pd.read_csv('dataX.txt')
display(data)
```

```python
plt.figure(figsize=(8,5))
count = data['Nguon X'][2:]
ch = np.arange(2, 8192)

E = 0.2598*ch + 6.9849

plt.plot(E, count, linewidth=0.8, color='blue', label="Phổ thực nghiệm")
plt.xlim(-10, 1600)


xs = np.array([194,295,1057,1173,1376,1482,4573,5203])
xs_E = 0.2598*xs + 6.9849
ys = (2250,470,110,127+30,345+50,55+30,60,60)
plt.scatter(xs_E,ys, color='red', label="Đỉnh năng lượng")

plt.text(57+30, 2150-50, "E: 57" + "\n" + "ch: 194", fontsize=13)
plt.text(83+30, 470-40, "E: 83"+ "\n" + "ch: 295", fontsize=13)
plt.text(363+10, 345+90, "E: 363"+ "\n" + "ch: 1376", fontsize=13)
plt.text(393+30, 55+30, "E: 390"+ "\n" + "ch: 1482", fontsize=13)
plt.text(1191-220, 60+50, "E: 1191"+ "\n" + "ch: 4573", fontsize=13)
plt.text(1354-100, 55+100, "E: 1354"+ "\n" + "ch: 5203", fontsize=13)
plt.text(500, 1250, r"E [keV] = 0,2598.ch + 6,9849", fontsize=13)

plt.legend(fontsize=13)

plt.title("Phổ thực nghiệm nguồn X" + "\n" + "13 bit, 8192 kênh (ch)", fontsize=13)
plt.xlabel("Năng lượng keV", fontsize=13)
plt.ylabel("Số đếm", fontsize=13)

#plt.savefig("nguonX.png", dpi=300)
```

```python
import matplotlib as mpl
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt

ax = plt.figure(figsize=(11,13)).add_subplot(projection='3d')

# Plot a sin curve using the x and y axes.
cn = data['Nguon X'][2:5500]
k = np.arange(2, 5500)
e = 0.2598*k + 6.9849
ax.plot(k, e-30, cn, c='b', label="Phổ thực nghiệm")
#

# Scatter 3D
xs = np.array([194,295,1057,1173,1376,1482,4573,5203])
xs_E = 0.2598*xs + 6.9849
ys = (2250,470,110,127+30,345+50,55+30,60,60)
ax.scatter3D(xs, xs_E-30, ys, color='red', label="Đỉnh năng lượng",depthshade=False)
#
# Make legend, set axes limits and labels
legend = ax.legend(loc=0, bbox_to_anchor=(0.9, 0.8), fontsize=15)
legend.get_frame().set_alpha(1)

ax.set_xlabel('Kênh', fontsize=13, rotation=-37)
ax.set_ylabel('Năng lượng (keV)', fontsize=13)
ax.set_zlabel('Số đếm', fontsize=13)

# Customize the view angle so it's easier to see that the scatter points lie
# on the plane y=0
ax.set_xlim(0, 6000)
ax.set_ylim(0, 1500)
ax.set_zlim(0, max(cn) + 100)

ax.text(194+30, 57+30+30-30, 2150-50+50, "E: 57", fontsize=13)
ax.text(295+30, 83+30+30-30, 470-40, "E: 83", fontsize=13)
ax.text(1057-90, 280-180-30, 110+30+120, "E: 281", fontsize=13)
ax.text(1173-90, 310-120-30, 127+30+30+160, "E: 311", fontsize=13)
ax.text(1376+30, 363+10+30-30, 345+90, "E: 363", fontsize=13)
ax.text(1482+30, 393+30+30-30, 55+30, "E: 390", fontsize=13)
ax.text(4573-400, 1191, 60+50-30, "E: 1191", fontsize=13)
ax.text(5203, 1354-50, 55+100-30, "E: 1354", fontsize=13)
ax.text(2000, 400, 1500, r"E [keV] = 0,2598.ch + 6,9849", fontsize=15)
ax.view_init(elev=20., azim=-35, roll=0)

#plt.savefig("nguonX3d.png", dpi=300)
plt.show()
```
