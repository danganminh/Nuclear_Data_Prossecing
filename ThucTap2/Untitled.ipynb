# Ba-133

```python
ba_133 = pd.read_excel('Ba-133.xlsx')
```

```python
display(ba_133)
```

```python
y_ba = ba_133['Ba-133'][0:]
x_ba = np.linspace(1, len(y_ba), len(y_ba))
```

```python
plt.plot(x_ba, y_ba, linewidth=0.5, color='red')
plt.xlim([0, 500])
plt.title('Ba-133')
```

```python
# two peak energy
peak1_ba = max(y_ba[0:100]) # peak 1
peak2_ba = max(y_ba[200:400])  # peak 2

print('peak 1 is ' + str(peak1_ba)) 
print('peak 2 is ' + str(peak2_ba))

# show index + 1 == channel
def index_peak1(y_ba):
    for i in range(100):
        if y_ba[i] == max(y_ba[0:100]):
            return i+1
def index_peak2(y_ba):
    for i in range(200, 400):
        if y_ba[i] == max(y_ba[200:400]):
            return i+1

print(str(index_peak1(y_ba)) + ' is index of count ' + str(peak1_ba))
print(str(index_peak2(y_ba)) + ' is index of count ' + str(peak2_ba))
```

## 2) Trình bày kết quả chuẩn năng lượng theo kênh (trình bày đường chuẩn năng lượng và phương trình chuẩn năng lượng). Ở đó, phương trình chuẩn năng lượng có dạng E(keV) = a.kênh + b

### Ba-133

```python
plt.figure(figsize=[8, 5])

plt.plot(x_ba, y_ba, color='blue', label='Phổ')
plt.xlim([0, 400])

plt.title('Phổ Ba-133' + '\n' + 'Real Time: 323.00(s), Live Time: 300.00(s)', fontname='arial', fontsize=14)
plt.ylabel('Số Đếm', fontname='arial', fontsize=13)
plt.xlabel('Kênh (2048)', fontname='arial', fontsize=13)

#31 is index of count 31707
#288 is index of count 6246

plt.text(40, 29000, 'Kênh: 31' + '\n' + 'Số đếm: 31707', fontname='arial', fontsize=13)
plt.scatter(31, 31707, color='red', label='Đỉnh năng lượng')

plt.text(305, 6246, 'Kênh: 288' + '\n' + 'Số đếm: 6246', fontname='arial', fontsize=13)
plt.scatter(288, 6246, color='red')

plt.legend(fontsize=13)
```

```python
# peak 1 channel 31, E = 212.49 keV
# peak 2 channel 288, E = 19.6847 keV
# E = a.channel + b

plt.figure(figsize=[8, 5])

E_ba = [212.49, 19.6847]
channel_ba = [31, 288]
fit_ba = np.polyfit(channel_ba, E_ba, 1)
x_sm_ba = np.linspace(np.min(channel_ba), np.max(channel_ba), 1000)
E_ba_fit = fit_ba[0]*x_sm_ba + fit_ba[1]

plt.scatter(channel_ba, E_ba, color='red', label='Đỉnh năng lượng')
plt.plot(x_sm_ba, E_ba_fit, color='blue', label='Linear fit')
plt.xlim([10, 310])
plt.ylim([5, 230])

plt.title('Phương trình chuẩn năng lượng Ba-133', fontname='arial', fontsize=14)
plt.ylabel('Năng lượng $(keV)$', fontname='arial', fontsize=13)
plt.xlabel('Kênh (2048)', fontname='arial', fontsize=13)
plt.legend(fontsize=13)

plt.text(30, 90, r'$E =$ ' + "- " + str(abs(round(fit_ba[0],2))) + r'$\times$kênh + ' + str(round(fit_ba[1],2)) + r' $(keV)$'
         , fontsize=13, color='blue')
plt.text(50, 210, 'Kênh: 31 = 212.49 keV', fontname='arial', fontsize=12, color='red')
plt.text(180, 19.6847, 'Kênh: 288 = 19.6847 keV', fontname='arial', fontsize=12, color='red')
```
