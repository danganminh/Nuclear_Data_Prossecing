```python
plt.figure(figsize=[8, 5])

# Ba-133
# E =80.9979, 302.8508, 356.0129
# ch =66, 242, 287

# Cobalt-60
# E = 1173.228, 1332.492
# ch = 851, 962

# Eu-152
# E = 40.1186, 121.7817, 244.6974, 344.2785, 964.079, 1408.013, 1112.076, 778.9045
# ch = 31, 93, 183, 256, 702, 1017, 796, 567

e_sor = sorted([80.9979, 302.8508, 356.0129, 1173.228, 1332.492, 121.7817, 244.6974, 344.2785, 964.079, 1408.013, 1112.076, 778.9045])
ch_sor = sorted([66, 242, 287, 851, 962, 93, 183, 256, 702, 1017, 796, 567])

E = np.array([80.9979, 121.7817, 244.6974, 302.8508, 344.2785, 356.0129, 778.9045, 964.079, 1112.076, 1173.228, 1332.492, 1408.013])
ch = np.array([66, 93, 183, 242, 256, 287, 567, 702, 796, 851, 962, 1017])

fit = np.polyfit(ch, E, 1)
ch_sm = np.linspace(min(ch), max(ch), 1000)
yfit = fit[0]*ch_sm + fit[1]

#plt.scatter(ch, E, color='black', label='Data')
plt.plot(ch_sm, yfit, color='darkorange', label='Linear Fit')

y_in = fit[0]*ch + fit[1]
R2 = 1 - (sum((E-y_in)**2)) / (sum((E-np.mean(E))**2))
print("R^2 = " + str(round(R2,4)) )

print('E = ' + str(fit[0]) + 'x ' + str(fit[1]))

# scatter Ba-133
E_ba = [80.9979, 302.8508, 356.0129]
ch_ba = [66, 242, 287]
plt.scatter(ch_ba, E_ba, color='red', label='Ba-133')

# scatter Co-60
E_co = [1173.228, 1332.492]
ch_co = [851, 962]
plt.scatter(ch_co, E_co, color='blue', label='Co-60')

#scatter Eu-152
E_eu = [121.7817, 244.6974, 344.2785, 778.9045, 964.079, 1112.076, 1408.013]
ch_eu = [93, 183, 256, 567, 702, 796, 1017]
plt.scatter(ch_eu, E_eu, color='green', label='Eu-152')

plt.title('Phng trnh chun nng lng', fontname='arial', fontsize=14)
plt.xlabel('Knh', fontname='arial', fontsize=13)
plt.ylabel(r'Nng lng $(keV)$', fontname='arial', fontsize=13)
plt.legend(fontsize=13)
plt.xlim([-10, 1100])
plt.ylim([-10, 1500])

plt.text(580, 600, r'$E$ = ' + str(round(fit[0], 2)) + r'$\times$ knh - ' + str(abs(round(fit[1], 2))) + r' $(keV)$',
         fontsize=13, color='darkorange')
plt.text(580, 500, r'$R^2$ = ' + str(round(R2,4)), fontsize=13, color='darkorange')

#plt.savefig('Peak-Plot_cn.png', dpi=300)
```
