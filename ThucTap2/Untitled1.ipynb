```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
```

## Ba-133

```python
ba_133 = pd.read_excel('Ba-133.xlsx')
```

```python
y_ba = ba_133['Ba-133'][0:]
x_ba = np.linspace(0, len(y_ba), len(y_ba))
len(x_ba)
```

```python
plt.plot(x_ba, y_ba, linewidth=0.5, color='red')
plt.xlim([0, 500])
plt.title('Ba-133')
```

```python
# two peak energy
peak1_ba = max(y_ba[0:100]) # peak 1
peak2_ba = max(y_ba[200:400])  # peak 2

print('peak 1 is ' + str(peak1_ba)) 
print('peak 2 is ' + str(peak2_ba))

# show index + 1 == channel
def index_peak1(y_ba):
    for i in range(100):
        if y_ba[i] == max(y_ba[0:100]):
            return i+1
def index_peak2(y_ba):
    for i in range(200, 400):
        if y_ba[i] == max(y_ba[200:400]):
            return i+1

print(str(index_peak1(y_ba)) + ' is index of count ' + str(peak1_ba))
print(str(index_peak2(y_ba)) + ' is index of count ' + str(peak2_ba))
```

```python
plt.figure(figsize=[8, 5])

plt.plot(x_ba, y_ba, color="red", label='Ph')
plt.xlim([0, 400])

plt.title('Ph Ba-133' + '\n' + 'Real Time: 323.00(s), Live Time: 300.00(s)', fontname='arial', fontsize=14)
plt.ylabel('S m', fontname='arial', fontsize=13)
plt.xlabel('Knh (2048)', fontname='arial', fontsize=13)

#31 is index of count 31707
#288 is index of count 6246

plt.text(40, 29000, 'Knh: 31' + '\n' + 'S m: 31707', fontname='arial', fontsize=13)
plt.scatter(31, 31707, color='blue', label='nh nng lng')

plt.text(305, 6246, 'Knh: 288' + '\n' + 'S m: 6246', fontname='arial', fontsize=13)
plt.scatter(288, 6246, color='blue')

plt.legend(fontsize=13)
```
