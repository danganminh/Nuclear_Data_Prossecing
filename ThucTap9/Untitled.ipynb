```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_excel("Book1.xlsx")
print(data)
```

```python
channel = data["Vị trí kênh K"][0:]
energy = data["Năng lượng phát gamma\nE (keV)"][0:]
```

```python
fit = np.polyfit(channel, energy, 1)
yfit = fit[0]*channel + fit[1]
plt.plot(channel, yfit, c='b', label="Đường chuẩn năng lượng")
plt.scatter(channel, energy, c='r', label="Đỉnh năng lượng")
plt.text(3000, 500, r"$E = $"+ "0,26" + r"$K$ + " + "7,26", fontsize=13)
plt.legend()
plt.xlabel("Kênh K")
plt.ylabel("Năng lượng (keV)")
#plt.savefig("plot.png", dpi=300)
```
