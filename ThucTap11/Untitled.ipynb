```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import scipy
```

```python
data = pd.read_excel("NguonChuan.xlsx")
display(data)
```

```python
plt.figure(figsize=(8, 5))

k_a = np.array(data['K_alpha'][0:])
k_b = np.array(data['K_beta'][0:])

e_a = np.array(data['E_alpha'][0:])
e_b = np.array(data['E_beta'][0:])

K = np.append(k_a, k_b)
E = np.append(e_a, e_b)

k_sm = np.linspace(min(K), max(K), 1000)
fit = np.polyfit(K, E, 1)
yfit_sm = fit[0]*k_sm + fit[1]
yfit = fit[0]*K + fit[1]

plt.plot(k_sm, yfit_sm, color='crimson', label="ng chun")

kenh = []
for i in range(10):
    kenh.append(K[i])
    kenh.append(K[i+10])

nang_luong = []
for i in range(10):
    nang_luong.append(E[i])
    nang_luong.append(E[i+10])
    
k_1 = kenh[0:2]
E_1 = nang_luong[0:2]

k_2 = kenh[2:4]
E_2 = nang_luong[2:4]

k_3 = kenh[4:6]
E_3 = nang_luong[4:6]

k_4 = kenh[6:8]
E_4 = nang_luong[6:8]

k_5 = kenh[8:10]
E_5 = nang_luong[8:10]

k_6 = kenh[10:12]
E_6 = nang_luong[10:12]

k_7 = kenh[12:14]
E_7 = nang_luong[12:14]

k_8 = kenh[14:16]
E_8 = nang_luong[14:16]

k_9 = kenh[16:18]
E_9 = nang_luong[16:18]

k_10 = kenh[18:20]
E_10 = nang_luong[18:20]

plt.scatter(k_1, E_1, color = 'r', label="Ti")
plt.scatter(k_2, E_2, color = 'b', label="Mn")
plt.scatter(k_3, E_3, color = 'deeppink', label="Zn")
plt.scatter(k_4, E_4, color = 'pink', label="Br")
plt.scatter(k_5, E_5, color = 'orange', label="Co")
plt.scatter(k_6, E_6, color = 'yellow', label="V")
plt.scatter(k_7, E_7, color = 'brown', label="As")
plt.scatter(k_8, E_8, color = 'cyan', label="Cr")
plt.scatter(k_9, E_9, color = 'lightsteelblue', label="Fe")
plt.scatter(k_10, E_10, color = 'darkviolet', label="Ni")
plt.legend(fontsize=11)
plt.xlabel("Knh K", fontsize=13)
plt.ylabel("E (keV)", fontsize=13)

print(fit[0])
print(fit[1])

def rsquared(x, y):
    """ Return R^2 where x and y are array-like."""
    slope, intercept, r_value, p_value, std_err = scipy.stats.linregress(x, y)
    return r_value**2
R = rsquared(K, E)
print(round(R, 4))

plt.text(800, 8, r"$ E = 0,0125K - 0,0152$", fontsize=12)
plt.text(800, 7.25, r"$ R^2 = 0,9999$", fontsize=12)

#plt.savefig('plot.png', dpi=300)
```
